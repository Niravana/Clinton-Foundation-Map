```python
import pandas as pd 
import folium
import random

m = folium.Map(location = [40.0150, -105.2705],zoom_start=6,tiles='Mapbox bright')


def df_city():
    counter = 0
    x=[]
    y=[]
    while counter <=50:
        counter += 1
        x.append(random.uniform(31.45,41.35))
        y.append(random.uniform(-120,-85))
    return x,y
x,y = df_city()

for a,b in zip(x,y):
    folium.Marker(location = [a, b], popup = 'City that we helped!').add_to(m)
m
```
